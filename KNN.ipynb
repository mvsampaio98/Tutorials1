```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Marcelo Sampaio/Desktop/Datasets/cancer.csv')
df.head()
```

```python
del df['id']
del df['Unnamed: 32']
```

```python
df['diagnosis'].unique()
```

```python
# M=0 and B=1
idCat, cat = pd.factorize(df['diagnosis'])
```

```python
idCat
```

```python
df['diagnosisCat'] = idCat
del df['diagnosis']
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop('diagnosisCat',axis=1)
y = df['diagnosisCat']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Padronizando com Standard Scaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = sc.transform(X_test)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
# Definir HiperparÃ¢metros
params = [
    {
        'n_neighbors':[1,2,6,9,12,15],
        'weights':['uniform','distance']
    }
]
ins = KNeighborsClassifier(n_neighbors=6, weights= 'distance')
grid_search = GridSearchCV(ins,params,cv=5)
grid_search.fit(X_train,y_train)
grid_search.best_params_
```

```python
ins.fit(X_train,y_train)
pred= ins.predict(X_test)
pred[:10]
```

```python
y_test[:10]
```

```python
ins.score(X_test,y_test)
```

```python
cross = cross_val_score(ins,X_test,y_test,cv=10,scoring='accuracy')
final = sum(cross)/len(cross)
final
```
