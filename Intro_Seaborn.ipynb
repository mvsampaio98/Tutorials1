**Intro to Seaborn** 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/Marcelo Sampaio/Desktop/Datasets/flights.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.displot(data['passengers'], bins=50)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='year',y='passengers',data=data)
```

```python
sns.jointplot(x='year',y='passengers',data=data)
```

```python
plt.figure(figsize=(12,8))
sns.pairplot(data)
```

```python
sns.heatmap(data.corr(), annot=True)
```
