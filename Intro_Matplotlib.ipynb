**Matplotlib with DataFrame**

```python
data_netflix = pd.read_csv('C:/Users/Marcelo Sampaio/Desktop/Datasets/netflix_titles.csv')
data_netflix.head()
```

```python
x = data_netflix['release_year'].value_counts().index
y = data_netflix['release_year'].value_counts().values
```

```python
fig, axs = plt.subplots(figsize=(15,7))
axs.set_xlabel('Years')
axs.set_ylabel('Number of movies')
axs.set_title('Number of movies produced per year')
axs.bar(x,y)
```

```python
x = data_netflix['type'].value_counts().index
y = data_netflix['type'].value_counts().values
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(y,labels=x,autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```
